```python
import keras.preprocessing.image as k_image
```

```python
from VGG19_cheater import ClassCheater
from visualization import generate_heatmap, plot_heatmap, plot_heatmap_cover
```

## Simple example: tiger to jaguar

```python
# tiger = 292, jaguar = 290
cheater = ClassCheater('images/tiger.jpg', 290)
```

#### Before changes

```python
cheater.get_image()
```

```python
cheater.get_predictions()
```

```python
heatmap = generate_heatmap(cheater.model, cheater.x)
plot_heatmap(heatmap, blur=5)
plot_heatmap_cover(heatmap, k_image.img_to_array(cheater.image) / 255, blur=5, cutoff=0.35)
```

#### Cheating

```python
cheater.optimize()
```

#### After

```python
cheater.get_image()
```

```python
cheater.get_predictions()
```

```python
heatmap = generate_heatmap(cheater.model, cheater.x)
plot_heatmap(heatmap, blur=5)
plot_heatmap_cover(heatmap, k_image.img_to_array(cheater.image) / 255, blur=5, cutoff=0.35)
```

```python
cheater.get_image().save('outputs/tiger-jaguar.jpg')
```
